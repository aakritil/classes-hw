```
import matplotlib.pyplot as plt

k = 500
m = 1.67*10**-27
x = 2.0*10**-12
v= 0
dt = 1*10**-15
t=0.0
x_list = []
t_list = []
velocity_list =[]
KE_sum = 0.0
Ke_list = []

for i in range(351):
  a = -k*x/m
  x += v*dt + 0.5*a*dt**2 
  anext = -k * x / m
  v += ((a + anext) / 2) * dt
  KE = .5*m*v**2
  KE_sum += KE
  t += dt
  x_list.append(x)
  Ke_list.append(KE)
  velocity_list.append(v)
  t_list.append(t)

sumavg = KE_sum/len(Ke_list)
temp = (2*sumavg)/(1.38*10**-23)

print(temp)

plt.figure()
plt.title('Movement')
plt.xlabel('Time (seconds)')
plt.ylabel('Location (meters)')
plt.scatter(t_list,x_list)

plt.figure()
plt.title('Kinetic Energy')
plt.xlabel('Time (seconds)')
plt.ylabel('Energy (joules)')
plt.scatter(t_list,Ke_list)

```
